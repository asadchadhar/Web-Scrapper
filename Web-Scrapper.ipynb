```python
import requests
from bs4 import BeautifulSoup


request = requests.get('https://tribune.com.pk/')

response = request.text

soup = BeautifulSoup(response,'html.parser')

for x in soup.findAll('span', class_ ='bullets-story-title'):
        
        news_title = x.contents[0]
        
        print(news_title)
```
